```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Churn_Modelling.csv")
```

```python
df.head(3)
```

```python
df=df.drop(columns=["RowNumber","CustomerId","Surname"])
```

```python
df=pd.get_dummies(df,columns=["Geography","Gender"],drop_first=True)
```

```python
df.head(3)
```

```python
y=df['Exited'] # to be predicted
x=df.drop("Exited",axis=1) # the parameters required to estimate
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=61,test_size=0.7)
```

```python
rf=RandomForestClassifier()

model=rf.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[712,31,9,0,5,1,1,200000,False,False,True]])
```

# One Hot Encoding Method

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Churn_Modelling.csv")
df=df.drop(columns=["RowNumber","CustomerId","Surname"])
```

```python
ohe=OneHotEncoder()
xd=ohe.fit_transform(df[["Geography","Gender"]]).toarray()
```

```python
ohe.get_feature_names_out()
```

```python
xd=pd.DataFrame(xd)
xd.columns=ohe.get_feature_names_out()
xd
```

```python
df=df.drop(columns=["Geography","Gender"])
```

```python
df[xd.columns]=xd
df.head(3)
```

```python
y=df['Exited']
x=df.drop("Exited",axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier()
model=tree.fit(x,y)
model.score(x,y)

# Overfitting
```
